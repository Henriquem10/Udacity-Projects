got the global temperature and the year. One thing of importance is that the Rio De Janeiro data doesn't have any temperature recorded between 1844 to 1850 so when taking the moving average for 7 years one of the results will end up with 0.

```python
select year from city_data
where city = 'Rio De Janeiro';

select year from global_data
```

Doing this query also helps to find out that the global data stored starts from the year 1750 to 2015 while the data in Rio starts from 1832 to 2013

```python
SELECT city_data.year,
    city_data.avg_temp as city_temp,
    global_data.avg_temp as global_temp
FROM city_data, global_data
WHERE city_data.year = global_data.year
    AND NOT city_data.avg_temp IS NULL
    AND city_data.city = 'Rio De Janeiro'
```

another way to get the data that you want for both Rio De Janeiro and Global while excluding the empty years

```python
import matplotlib.pyplot as plt
import pandas as pd

global_temp = pd.read_csv('tempGlobal.csv',sep=';')
city_temp = pd.read_csv('tempRio.csv',sep=';')
```

```python
MAglobal = global_temp['7-yearMA']
MArio= city_temp['7-yearMA']
yearGlobal = global_temp['year']
yearRio = city_temp['year']
```

```python
plt.title('Global temperature MA')
plt.plot(yearGlobal,MAglobal)
plt.xlabel('Year')
plt.ylabel('Moving Average')
plt.show()

plt.title('Rio De Janeiro temperature MA')
plt.plot(yearRio,MArio)
plt.xlabel('Year')
plt.ylabel('Moving Average')
plt.show()
```

Now the graphs but both starting at the year 1850

```python
plt.title('Global temperature MA')
plt.plot(yearGlobal,MAglobal)
plt.xlabel('Year')
plt.ylabel('Moving Average')
plt.xlim([1850, 2015])
plt.show()

plt.title('Rio De Janeiro temperature MA')
plt.plot(yearRio,MArio)
plt.xlabel('Year')
plt.ylabel('Moving Average')
plt.xlim([1851, 2015])
plt.show()
```

The first thing of notice beside the sudden spike in temperature in Rio because of the lack of data is that the global Temperature is way lower than the one in the city.

Following the global trend the city temperature in the last years has also been slightly increasing.

Also following the global trend during 1880 there was also a slightly drop in the temperature in the city.
